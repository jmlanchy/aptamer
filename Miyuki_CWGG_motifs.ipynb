```python
import pandas
from Bio import SeqIO
from Bio import motifs
from Bio.Seq import Seq

# CWGG or CWG

miyumotifs = (Seq("CAG"),Seq("CTG"),Seq("CAGG"),Seq("CTGG"))

for seq_record in SeqIO.parse("aptamers.fasta", "fasta"):
    for mot in miyumotifs:
        instances = [mot]
        m = motifs.create(instances)
        total = 0
        for seq in m.instances.search(seq_record.seq):
            #print(seq_record.name + " : %i : %s" %(pos,seq.transcribe()))
            total = total + 1
        print(str(seq_record.name) + "," + str(total) + "," + str(mot.transcribe()))
```
